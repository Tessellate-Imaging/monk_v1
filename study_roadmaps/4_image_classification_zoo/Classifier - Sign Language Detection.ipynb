<h2> Setting up the library,to perform classification</h2>

```
#Using pytorch backend 

# When installed using pip
from monk.pytorch_prototype import prototype
```

```
gtf = prototype(verbose=1);
gtf.Prototype("Project-SignLangMNIST", "Sign-lang-classification-vgg16");
```

```
gtf.Default(dataset_path="/content/Sign_lang_data/Train",
            path_to_csv="/content/Sign_lang_data/train_labels.csv",
            model_name="vgg16", 
            num_epochs=5);
```

<h2>Training the model</h2>

```
#Start Training
gtf.Train();
```

<h1> Validating and testing the model performance</h1>

<code>- Set flag eval_infer as True</code>

```
gtf = prototype(verbose=1);
gtf.Prototype("Project-SignLangMNIST", "Sign-lang-classification-vgg16", eval_infer=True);
```

```
gtf.Dataset_Params(dataset_path="/content/Sign_lang_data/Test",
            path_to_csv="/content/Sign_lang_data/test_labels.csv");
gtf.Dataset();
```
