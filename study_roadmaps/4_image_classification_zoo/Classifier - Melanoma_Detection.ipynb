#Melanoma Detection using Monk AI library

Melanoma of the skin is the 19th most commonly occurring cancer in men and women. There were nearly 300,000 new cases in 2018. The top 20 countries with the highest rates of melanoma of the skin in 2018.

Here we will try to detect Skin Cancer (Malignant) lesions, using Transfer Learning. I had previously experimented with models, like

Resnet50 : 79% accuracy <br>
VGG16 : 82% accuracy<br>
Resnext101_32x8d : 82% accuracy <br>

And finally with Alexnet, (86% accuracy). So in this project, I have used the Alexnet model.

## Setting up the library,to perform classification, from start

````
from monk.pytorch_prototype import prototype
````

````
gtf = prototype(verbose=1);
gtf.Prototype("Skin-Cancer-Detection", "sample-experiment-1");
````

## Training on Alexnet model

About the Alexnet model:

image.png

The Alexnet contains, 5 convolutional layers, and 3 fully connected layers. 

````
gtf.Default(dataset_path="/content/gdrive/My Drive/data/train/", 
            model_name="alexnet", 
            freeze_base_network=False,
            num_epochs=5);
````

The model is now loaded, and our data is split into training and validation sets with a 0.7 split-ratio.

````
#Start Training
gtf.Train();
````

image.png

## Validating and testing the model performance

<code>- Set flag eval_infer as True</code>

````
gtf = prototype(verbose=1);
gtf.Prototype("Skin-Cancer-Detection", "sample-experiment-1", eval_infer=True);
````

````
gtf.Dataset_Params(dataset_path="/content/gdrive/My Drive/data/test");
gtf.Dataset();
````

````
accuracy, class_based_accuracy = gtf.Evaluate();
````
